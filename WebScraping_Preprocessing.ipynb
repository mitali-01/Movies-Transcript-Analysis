```python
import requests
import pickle 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# Scrapes transcript data from scrapsfromtheloft.com
def url_to_transcript(url):
  page=requests.get(url)
  soup=BeautifulSoup(page.content, 'html5lib')
  text=soup.find_all('p')
  script=""
  for t in text:
    script=script+' '+t.get_text()
  return script
```

```python
urls=['https://scrapsfromtheloft.com/movies/black-panther-wakanda-forever-transcript/',
      'https://scrapsfromtheloft.com/movies/puss-in-boots-last-wish-transcript/',
      'https://scrapsfromtheloft.com/movies/minions-the-rise-of-gru-transcript/',
      'https://scrapsfromtheloft.com/movies/doctor-strange-in-the-multiverse-of-madness-transcript/',
      'https://scrapsfromtheloft.com/movies/shershaah-transcript/',
      'https://scrapsfromtheloft.com/movies/the-devil-all-the-time-transcript/',
      'https://scrapsfromtheloft.com/movies/knives-out-2019-script/',
      'https://scrapsfromtheloft.com/movies/the-boss-baby-family-business-transcript/',
      'https://scrapsfromtheloft.com/movies/black-widow-2021-transcript/',
      'https://scrapsfromtheloft.com/movies/the-conjuring-3-the-devil-made-me-do-it-transcript/']
movies=['Black Panther 2', 'Puss In Boots 2', 'Minions 2', 'Doctor Strange 2', 'Shershaah', 'the Devil All the Time', 'Knives Out', 'The Boss Baby', 'Black Widow', 'Conjuring 3']
```

```python
# request transcripts (takes a few minutes to run)
transcripts=[]
for u in urls:
  transcripts.append(url_to_transcript(u))
```

```python
for i,c in enumerate(movies):
  with open('/content/drive/MyDrive/NLP_Assignment/transcripts/'+c+'.txt', 'wb') as file:
    pickle.dump(transcripts[i],file)
```

```python
# Load pickled files
data={}
for i,c in enumerate(movies):
  with open('/content/drive/MyDrive/NLP_Assignment/transcripts/'+c+'.txt', 'rb') as file:
    data[c]=pickle.load(file)
```

```python
#put data in a data frame
import pandas as pd
pd.set_option('max_colwidth',150)
pd.set_option('max_colwidth', 150)
df=pd.DataFrame([data]).transpose()
df.columns=['transcript']
df=df.sort_index()
df
```

```python
import re
import string
import contractions
```

```python
#Cleaning text round 1
def clean_text_1(text):
  text = text.lower() #lower case
  text = re.sub('\[.*?\]', '', text) #remove text in brackets
  # text = re.sub('[%s]' % re.escape(string.punctuation), '', text) #remove punctiuation
  text = re.sub('\w*\d\w*', '', text) #remove words containing numbers
  return text
```

```python
df=pd.DataFrame(df.transcript.apply(lambda x: clean_text_1(x)))
df.head()
```

```python
#Cleaning text round 2
def clean_text_2(text):
  # text = re.sub('[‘’“”…]', '', text)
  text = re.sub('\n', '', text)
  text = re.sub('♪', '', text)
  return text
```

```python
df=pd.DataFrame(df.transcript.apply(lambda x: clean_text_2(x)))
df.head()
```
